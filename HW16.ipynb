```python
%%sql 
SELECT distinct arrival_airport, aircraft_code FROM flights f
WHERE f.arrival_airport='AER'
```

1. Which fare conditions are available for flights from AER to SVO (i.e. from Adler/Sochi to Sheremetevo/Moscow)

```python
%%sql 
SELECT DISTINCT fare_conditions
FROM ticket_flights
WHERE flight_id IN (
    SELECT flight_id
    FROM flights
    WHERE departure_airport = 'AER' AND arrival_airport = 'SVO'
);
```

This query retrieves the distinct fare conditions for flights from AER to SVO

2. A list of aircraft models which fly to the Yakutsk AND Abakan cities

```python
%%sql 
SELECT DISTINCT aircraft_code
FROM flights
WHERE arrival_airport = 'YKS'
AND aircraft_code IN (
    SELECT aircraft_code
    FROM flights
    WHERE arrival_airport = 'ABA'
);
```

This query filters aircraft models that fly to both Yakutsk (YKS) and Abakan (ABA) by ensuring that each aircraft code appears for both cities

3. A list of aircraft models which fly to the Yakutsk OR Abakan cities

```python
%%sql 
SELECT DISTINCT aircraft_code
FROM flights
WHERE arrival_airport = 'YKS' OR arrival_airport = 'ABA';
```

This query retrieves aircraft models that fly to either Yakutsk or Abakan

4. A list of aircrafts which DON’T fly to Yakutsk city

```python
%%sql 
SELECT DISTINCT aircraft_code
FROM flights
WHERE aircraft_code NOT IN (
    SELECT aircraft_code
    FROM flights
    WHERE arrival_airport = 'YKS'
);
```

This query finds aircraft codes that do not appear in flights arriving at Yakutsk

5. Find out airport codes related to Mrs. 'ELLA DMITRIEVA'  - from where and to where she flew.

```python
%%sql 
SELECT DISTINCT departure_airport, arrival_airport
FROM flights
WHERE flight_id IN (
    SELECT flight_id
    FROM ticket_flights
    WHERE ticket_no IN (
        SELECT ticket_no
        FROM tickets
        WHERE passenger_name = 'ELLA DMITRIEVA'
    )
);
```

This query retrieves unique departure and arrival airports for flights taken by Mrs. Ella Dmitrieva
